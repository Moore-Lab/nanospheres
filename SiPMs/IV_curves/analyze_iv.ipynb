```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = 'C:/data/SiPM'
#file_list = ['20240319_IVcurve_single_SiPM_board_double_box.csv',
#    '20240318_IVcurve_single_SiPM_board_dark_box.csv',
#             '20240318_IVcurve_single_SiPM_board.csv', 
#             '20240318_IVcurve_single_SiPM_board_taped_hole.csv',
#             '20240318_IVcurve_single_SiPM_board_arduino_unplugged.csv',
#             '20240319_IVcurve_single_SiPM_board_retaped.csv',
#             '20240319_IVcurve_single_SiPM_board_retaped_lightsoff.csv',
#             '20240319_IVcurve_single_SiPM_board_retaped_arduino_on.csv',]
#labels = ["Double box",'Metal box', 'No tape', 'With tape',"Arduino unplugged","Re-taped","Lights off","Arduino on"]

file_list = ['20240319_IVcurve_single_SiPM_board_double_box.csv',
             '20240403_IVcurve_coated_box_closed.csv',
             '20240404_IVcurve_coated.csv',
             '20240405_IVcurve_coated_alpha_foil.csv']
labels = ["Double box",'Box closed','Coated scint',"Coated + alpha foil"]
```

```python
plt.figure()

for i,f in enumerate(file_list):

    cdat = np.loadtxt(data_path + '/' + f, delimiter=',')
    gpts = (cdat[:,1] > 5e-10) &  (cdat[:,0] > 0) # fix issue at range switch

    if(labels[i] == "Double box"):
        plt.semilogy(cdat[gpts,0], cdat[gpts,1], label=labels[i], lw=3, color='k')
    else:
        plt.semilogy(cdat[gpts,0], cdat[gpts,1], label=labels[i])

plt.legend(fontsize=8)
plt.xlabel("Bias voltage [V]")
plt.ylabel("Current [A]")
plt.grid(True, which='both', axis='both')
plt.xlim(20,30)
```

```python
plt.figure()

for i,f in enumerate(file_list[-1:]):

    cdat = np.loadtxt(data_path + '/' + f, delimiter=',')
    gpts = cdat[:,1] > 5e-10 # fix issue at range switch

    if(labels[i] == "Double box"):
        plt.semilogy(cdat[gpts,0], cdat[gpts,1], label=labels[i], lw=3, color='k')
    else:
        plt.semilogy(cdat[gpts,0], cdat[gpts,1], label=labels[i])

yy = plt.ylim()
vbd = 24.7
plt.plot([vbd, vbd], yy, 'b:', label='$V_{bd}$ = %.1f'%vbd)
plt.ylim(yy)

plt.legend()
plt.xlabel("Bias voltage [V]")
plt.ylabel("Current [A]")
```
